```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

mouse_metadata_path = "..../data/Mouse_metadata.csv"
study_results_path = "..../data/Study_results.csv"

mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

data_complete = pd.merge(mouse_metadata, study_results, how="right" ,   on=["Mouse ID","Mouse ID"])
                        
data_complete.head()
```

```python
# the number of unique mice IDs in the data
data_complete["Mouse ID"].value_counts()
```

```python
# the number of unique mice IDs in the data

total_mice_id= len(data_complete["Mouse ID"].value_counts())
total_mice_id
```

```python
# check for any mouse ID with duplicate for Mouse ID & time points

Mouse_ID_duplicate = data_complete.loc[data_complete.duplicated(subset=["Mouse ID","Timepoint"]),"Mouse ID"].unique()
Mouse_ID_duplicate                                  
```

```python
# All the data for duplicate Mouse ID

duplicate= data_complete.loc[data_complete["Mouse ID"]=="g989"]
duplicate

```

```python
#  Create a clean DataFrame by dropping the duplicate mouse by its ID
# clean_df= data_complete.drop_duplicates(subset=["Mouse ID"],keep="first")
# clean_df.head()

clean_df= data_complete[data_complete["Mouse ID"].isin(duplicate)==False]
clean_df.head()
```

```python
# Checking the number of mice in the clean DataFrame

mice_clean_total= len(clean_df["Mouse ID"].value_counts())
mice_clean_total
```
